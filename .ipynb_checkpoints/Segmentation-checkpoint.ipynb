```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime, date
```

```python
seg_data = pd.read_excel("segmentation.xlsx")
```

```python
seg_data.sort_values("amount")
```

```python
ss1_20 = seg_data["amount"][(seg_data["amount"] >= 1) & (seg_data["amount"] <= 50)]
ss21_40 = seg_data["amount"][(seg_data["amount"] >= 51) & (seg_data["amount"] <= 100)]
ss41_60 = seg_data["amount"][(seg_data["amount"] >= 101) & (seg_data["amount"] <= 300)]
ss61_80 = seg_data["amount"][(seg_data["amount"] >= 301) & (seg_data["amount"] <= 600)]
ss81_100 = seg_data["amount"][(seg_data["amount"] >= 601)]

ssx = ["1-50", "51-100", "101-300", "301-600", "601->"]
ssy = [len(ss1_20.values), len(ss21_40.values), len(ss41_60.values), len(ss61_80.values), len(ss81_100.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=ssx, y=ssy, palette="nipy_spectral_r")
plt.title("Spending Scores")
plt.xlabel("Amount Paid")
plt.ylabel("Number of Customer")
plt.show()
```

```python
ssF = seg_data["gender"][(seg_data["gender"] == "F")]
ssM = seg_data["gender"][(seg_data["gender"] == "M")]

ssx = ["F", "M"]
ssy = [len(ssF.values), len(ssM.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=ssx, y=ssy, palette="nipy_spectral_r")
plt.title("Number of Passengers in Each Gender")
plt.xlabel("Count")
plt.ylabel("Gender")
plt.show()
```

```python
def age(born):
    born = born.date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
  
seg_data['Age'] = seg_data['dateOfBirth'].apply(age)
seg_data
```

```python
age18_25 = seg_data.Age[(seg_data.Age <= 25) & (seg_data.Age >= 18)]
age26_35 = seg_data.Age[(seg_data.Age <= 35) & (seg_data.Age >= 26)]
age36_45 = seg_data.Age[(seg_data.Age <= 45) & (seg_data.Age >= 36)]
age46_55 = seg_data.Age[(seg_data.Age <= 55) & (seg_data.Age >= 46)]
age55above = seg_data.Age[seg_data.Age >= 56]

x = ["18-25","26-35","36-45","46-55","55+"]
y = [len(age18_25.values),len(age26_35.values),len(age36_45.values),len(age46_55.values),len(age55above.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=x, y=y, palette="rocket")
plt.title("Number of Customer and Ages")
plt.xlabel("Age")
plt.ylabel("Number of Customer")
plt.show()
```

```python
PRESENT = datetime.today()
rfm = seg_data.groupby('contactID').agg({'paymentDate': lambda date: (PRESENT - date.max()).days,
                                        'paymentID': lambda num: count(num),
                                        'amount': lambda price: price.sum()})
```

```python
rfm
```

```python
rfm.columns=['recency','frequency','monetary']
rfm['recency'] = rfm['recency'].astype(int)
rfm
```

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'].rank(method='first'), 4 , ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'].rank(method='first'), 4 , ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'].rank(method='first'),4 , ['4','3','2','1'])
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.sort_values('RFM_Score', ascending=False)
```
